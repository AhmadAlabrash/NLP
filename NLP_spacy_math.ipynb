```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp('bread sandwich burger car tiger human wheat')
```

```python
for word in doc:
  print(word.text ,'||',word.vector)
```

```python
def similar(w1,w2):
  nw1 = nlp(w1)
  nw2 =nlp(w2)

  for w1 ,w2 in zip(nw1,nw2):
    print(w1.text ,'||' ,w2.text ,'||', w1.similarity(w2))
```

```python
similar('apple yahoo galaxy usa Elon','nokia google chokolate china Mars')
```
